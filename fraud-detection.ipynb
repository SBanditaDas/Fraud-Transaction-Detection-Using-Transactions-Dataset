```python
import numpy as np
import pandas as pd
import os
```

```python
data_path = '/kaggle/input/fraud-detailed-dataset/data/'
```

```python
pkl_files = [os.path.join(data_path, f) for f in os.listdir(data_path) if f.endswith('.pkl')]
print("Found files:", pkl_files)
```

```python
df_list = [pd.read_pickle(file) for file in pkl_files]
fraud_df = pd.concat(df_list, ignore_index=True)
```

```python
print("Shape:", fraud_df.shape)
print("Columns:", fraud_df.columns)
fraud_df.head()
fraud_df.info()
fraud_df.describe()
fraud_df['TX_FRAUD'].value_counts()
```

```python
features = ['TX_AMOUNT', 'CUSTOMER_ID', 'TERMINAL_ID']  # You can expand this later with engineered features
target = 'TX_FRAUD'

X = fraud_df[features]
y = fraud_df[target]

print("Feature matrix shape:", X.shape)
print("Target vector shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
train_preds = model.predict(X_train)
print("Training Accuracy:", accuracy_score(y_train, train_preds))
```

```python
test_preds = model.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, test_preds))
print("\nClassification Report:\n", classification_report(y_test, test_preds))
print("Confusion Matrix:\n", confusion_matrix(y_test, test_preds))
```

```python
input_data = {
    'TX_AMOUNT': 250,
    'CUSTOMER_ID': 12345,
    'TERMINAL_ID': 67890
}

input_df = pd.DataFrame([input_data])

prediction = model.predict(input_df)
probability = model.predict_proba(input_df)[0][1]

if prediction[0] == 1:
    print(f"⚠️ Fraudulent Transaction Detected! (Confidence: {probability:.2f})")
else:
    print(f"✅ Legitimate Transaction (Confidence: {1 - probability:.2f})")
```
